```python
df['Reviewer_Score'] = pd.to_numeric(df['Reviewer_Score'], errors='coerce')
df.dropna(subset=['Reviewer_Score'], inplace=True)

df['Calc_Average_Score'] = df.groupby('Hotel_Name')['Reviewer_Score'].transform('mean').round(1)
df['Calc_Average_Score']
```

```python
df['Average_Score'] = pd.to_numeric(df['Average_Score'], errors='coerce')
df.dropna(subset=['Average_Score'], inplace=True)

hotels= df['Hotel_Name'].value_counts()
count = 0

for k in hotels.index:
  df_filtered = df[df['Hotel_Name'] == k]
  Average_Score = df_filtered['Average_Score'].nlargest(1)
  Calc = df_filtered['Calc_Average_Score'].nlargest(1)
  if Average_Score.iloc[0] == Calc.iloc[0]:
    count += 1
count
```

```python
no_negative_reviews = df.apply(lambda x: True if x['Negative_Review'] == "No Negative" else False , axis=1)
print("Number of No Negative reviews: " + str(len(no_negative_reviews[no_negative_reviews == True].index)))

no_positive_reviews = df.apply(lambda x: True if x['Positive_Review'] == "No Positive" else False , axis=1)
print("Number of No Positive reviews: " + str(len(no_positive_reviews[no_positive_reviews == True].index)))

both_no_reviews = df.apply(lambda x: True if x['Negative_Review'] == "No Negative" and x['Positive_Review'] == "No Positive" else False , axis=1)
print("Number of both No Negative and No Positive reviews: " + str(len(both_no_reviews[both_no_reviews == True].index)))
```

```python
df.drop(['lat'], axis=1, inplace=True)
df.drop(['lng'], axis=1, inplace=True)
df.drop(['Review_Total_Negative_Word_Counts'], axis=1, inplace=True)
df.drop(['Review_Total_Positive_Word_Counts'], axis=1, inplace=True)
df.head()
```

```python
def replace_address(row):
    if "Netherlands" in row["Hotel_Address"]:
        return "Amsterdam, Netherlands"
    elif "Barcelona" in row["Hotel_Address"]:
        return "Barcelona, Spain"
    elif "United Kingdom" in row["Hotel_Address"]:
        return "London, United Kingdom"
    elif "Milan" in row["Hotel_Address"]:
        return "Milan, Italy"
    elif "France" in row["Hotel_Address"]:
        return "Paris, France"
    elif "Vienna" in row["Hotel_Address"]:
        return "Vienna, Austria"

df["Hotel_Address"] = df.apply(replace_address, axis = 1)
print(df["Hotel_Address"].value_counts())
```

```python
df.drop(['Additional_Number_of_Scoring'], axis=1, inplace=True)
df.drop(['Review_Date'] , axis=1, inplace=True)
df.drop(['Average_Score'] , axis=1, inplace=True)
```

```python
df.Total_Number_of_Reviews = df.groupby('Hotel_Name').Reviewer_Score.value_counts()
df.Average_Score = round(df.groupby('Hotel_Name').Reviewer_Score.transform('mean'),1)
df.head()
```

```python
df.Tags = df.Tags.str.strip("[']")
df.Tags = df.Tags.str.replace(" ', '", ",", regex = False)
```

```python
tag_split = df['Tags'].str.split(',', expand=True).iloc[:, :6]

tag_split = tag_split.apply(lambda col: col.str.strip())

all_tags = tag_split.stack()

tag_counts = all_tags.value_counts()

print(tag_counts)
print(f"\nTotal unique tags: {tag_counts.shape[0]}")
```

```python
filtered_tags = all_tags[~all_tags.str.contains(r'^Stayed \d+ night[s]?', case=False, na=False)]
filtered_tags = filtered_tags[~filtered_tags.str.contains(r'\broom[s]?\b', case=False, na=False)]
filtered_tags = filtered_tags[~filtered_tags.str.contains(r'^Submitted from a mobile device[s]?\b', case=False, na=False)]
filtered_tags = filtered_tags[~filtered_tags.str.contains(r'\bsuite[s]?\b', case=False, na=False)]
filtered_tags = filtered_tags[~filtered_tags.str.contains(r'\bapartment[s]?\b', case=False, na=False)]
filtered_tags = filtered_tags[~filtered_tags.str.contains(r'\bhotel[s]?\b', case=False, na=False)]
unique_tags = filtered_tags.unique().tolist()
print(unique_tags)
```

```python
# Apply the lambda function and reset the index to match the original DataFrame index
# Modified approach to create indicator columns directly from the tag_split DataFrame

# Instead of filtering and then applying, create the columns based on whether the original
# tags in the tag_split DataFrame contain the specific keywords.
df["Leisure_trip"] = tag_split.apply(lambda col: col.str.contains("Leisure trip", na=False)).any(axis=1).astype(int)
df["Couple"] = tag_split.apply(lambda col: col.str.contains("Couple", na=False)).any(axis=1).astype(int)
df["Solo_traveler"] = tag_split.apply(lambda col: col.str.contains("Solo traveler", na=False)).any(axis=1).astype(int)
df["Business_trip"] = tag_split.apply(lambda col: col.str.contains("Business trip", na=False)).any(axis=1).astype(int)
df["Group"] = tag_split.apply(lambda col: col.str.contains("Group|Travelers with friends", regex=True, na=False)).any(axis=1).astype(int)
df["Family_with_young_children"] = tag_split.apply(lambda col: col.str.contains("Family with young children", na=False)).any(axis=1).astype(int)
df["Family_with_older_children"] = tag_split.apply(lambda col: col.str.contains("Family with older children", na=False)).any(axis=1).astype(int)
df["With_a_pet"] = tag_split.apply(lambda col: col.str.contains("With a pet", na=False)).any(axis=1).astype(int)

df.head()
```

```python
df.drop(["days_since_review", "Total_Number_of_Reviews_Reviewer_Has_Given"], axis = 1, inplace=True)
```

```python
import time
import pandas as pd
import nltk as nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
nltk.download('stopwords')
cache = set(stopwords.words('english'))
def remove_stopwords(review):
    text = " ".join([word for word in review.split() if word not in cache])
    return text

# Remove the stop words from both columns
df.Negative_Review = df.Negative_Review.apply(remove_stopwords)
df.Positive_Review = df.Positive_Review.apply(remove_stopwords)
```

```python
df.head()
```

```python
vader_sentiment = SentimentIntensityAnalyzer()

def calc_sentiment(review):
    if review == "No Negative" or review == "No Positive":
        return 0
    return vader_sentiment.polarity_scores(review)["compound"]


df["Negative_Sentiment"] = df.Negative_Review.apply(calc_sentiment)
df["Positive_Sentiment"] = df.Positive_Review.apply(calc_sentiment)
```

```python
df.head()
```

```python
df = df.sort_values(by=["Negative_Sentiment"], ascending=True)
print(df[["Negative_Review", "Negative_Sentiment"]])
df = df.sort_values(by=["Positive_Sentiment"], ascending=True)
print(df[["Positive_Review", "Positive_Sentiment"]])

df = df.reindex(["Hotel_Name", "Hotel_Address", "Total_Number_of_Reviews", "Calc_Average_Score", "Reviewer_Score", "Negative_Sentiment", "Positive_Sentiment", "Reviewer_Nationality", "Leisure_trip", "Couple", "Solo_traveler", "Business_trip", "Group", "Family_with_young_children", "Family_with_older_children", "With_a_pet", "Negative_Review", "Positive_Review"], axis=1)
df.head()
```
